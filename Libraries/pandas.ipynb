```python
#all imp methods
import pandas as pd

# Create a sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [24, 27, 22, 32],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']
}
df = pd.DataFrame(data)

# Display the DataFrame
print("DataFrame:")
print(df)

# 1. Basic Info
print("\nBasic Info:")
print(df.info())  # Info about the DataFrame
print(df.describe())  # Statistical summary of numerical columns

# 2. Accessing Data
print("\nAccessing Data:")
print("First Row:")
print(df.loc[0])  # Access by label
print("First Row (using iloc):")
print(df.iloc[0])  # Access by position

# 3. Selecting Columns
print("\nSelecting Columns:")
print(df['Name'])  # Select a single column
print(df[['Name', 'Age']])  # Select multiple columns

# 4. Filtering Rows
print("\nFiltering Rows:")
print(df[df['Age'] > 25])  # Filter rows based on condition

# 5. Adding Columns
print("\nAdding Columns:")
df['Salary'] = [70000, 80000, 60000, 90000]  # Add a new column
print(df)

# 6. Dropping Columns
print("\nDropping Columns:")
df_dropped = df.drop(columns=['Salary'])  # Drop a column
print(df_dropped)

# 7. Handling Missing Values
print("\nHandling Missing Values:")
df_missing = df.copy()
df_missing.loc[1, 'Age'] = None  # Introduce a missing value
print(df_missing)
print("After Filling Missing Values:")
print(df_missing.fillna(df_missing.mean()))  # Fill missing values with the mean

# 8. Grouping Data
print("\nGrouping Data:")
grouped = df.groupby('City').mean()  # Group by city and calculate the mean
print(grouped)

# 9. Sorting Data
print("\nSorting Data:")
print(df.sort_values(by='Age', ascending=False))  # Sort by Age

# 10. Saving to CSV
df.to_csv('output.csv', index=False)  # Save DataFrame to CSV
print("\nDataFrame saved to 'output.csv'")


```
